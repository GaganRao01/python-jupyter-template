### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Data
    x = np.arange(0, 10)
    y = x**2
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]

    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    # First subplot - Line chart
    axs[0, 0].plot(x, y)
    axs[0, 0].set_title('Line Chart')

    # Second subplot - Bar chart
    axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
    axs[0, 1].set_title('Bar Chart')

    # Third subplot - Scatter plot
    axs[1, 0].scatter(x, y)
    axs[1, 0].set_title('Scatter Plot')

    # Fourth subplot - Pie chart
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    axs[1, 1].set_title('Pie Chart')

    plt.tight_layout()
    plt.show()

except ImportError as e:
    print(f"Error: Required library not found - {e}")
except ValueError as e:
    print(f"Error: Invalid data - {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
